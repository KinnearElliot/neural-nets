```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Layer, BatchNormalization, Input,  Add

import pandas as pd
import numpy as np

EPOCHS = 10

model_file = 'model_files/my_model'

# create callback that saves models weights
checkpoint_filepath = '/tmp/current_model_checkpoint.ckpt'
model_callback = tf.keras.callbacks.ModelCheckpoint(filepath = checkpoint_filepath,
                                                   save_weights_only = False,
                                                   save_best_only = True,
                                                   monitor = ['accuracy'],
                                                   mode = 'max')

train_df = pd.read_csv('./data/train.csv')
print(train_df.head())

model = keras.Sequential([
    keras.layers.Dense(4, input_shape=(2,), activation='relu'),
    keras.layers.Dense(2)])

model.compile(optimizer='adam',
             loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])

model.summary()

x = np.column_stack((train_df.x.values, train_df.y.values))

model.fit(x, train_df.color.values, batch_size=16, epochs = EPOCHS)

model.save(model_file)

# new_model = load_model(model_file)

# model.load_weights(checkpoint_filepath)

# ecg_length = 768
# signal = Input(shape=(ecg_length,1), dtype=np.float32, name='signal' )
```

```python
model = tf.keras.models.load_model(
    model_file, custom_objects=None, compile=False, options=None
)

model.compile(optimizer='adam',
             loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])


new_train_df = pd.DataFrame()
x_set = [1,2,3,4,5]
y_set = [2,3,4,5,6]
color = [0,0,0,1,1]

new_train_df['x'] = x_set
new_train_df['y'] = y_set
new_train_df['color'] = color

print(new_train_df.head())

model.summary()
x = np.column_stack((new_train_df.x.values, new_train_df.y.values))
model.fit(x, new_train_df.color.values, batch_size=16, epochs = EPOCHS)
```
